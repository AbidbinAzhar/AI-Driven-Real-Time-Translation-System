# Data Cleaning and Preprocessing

Cleaning

```python
#Removing punctuations
import re

def remove_punctuations(text):
  return re.sub(r'[^\w\s]', '', text)

df['title'] = df['title'].apply(remove_punctuations)
df['snippet'] = df['snippet'].apply(remove_punctuations)
```



1.   [ #Character block Start

```python
#Display data without punctuations
print(df)
```

```python
#Converting text to lowercase
import nltk

def to_lowercase(text):
    return text.lower()

# Apply the function to the title and snippet columns
df['title'] = df['title'].apply(to_lowercase)
df['snippet'] = df['snippet'].apply(to_lowercase)

print(df)
```

# Hugging face for translation

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

#Downloading the model and tokenizer
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")
```

```python
#Getting snippet based on title
def get_snippet(title):
    row = df[df['title'] == title]

    if not row.empty:
        return row['snippet'].values[0]
    else:
        return "Title not found."


title_to_search = 'stock market today what to watch  live updates'
snippet = get_snippet(title_to_search)
print(snippet)
```

```python
#Text to translate
text_en = snippet
```

```python
#Passing text on tokenizer and specifying Pytorch tensors
model_inputs = tokenizer(text_en, return_tensors='pt')
```

```python
#Translation into Hindi
generated_tokens_hi = model.generate(
    **model_inputs,
    forced_bos_token_id = tokenizer.lang_code_to_id["hi_IN"]
)
```

```python
translation_hi = tokenizer.batch_decode(generated_tokens_hi, skip_special_tokens=True)
translations = {
    "Hindi": translation_hi
}
```
